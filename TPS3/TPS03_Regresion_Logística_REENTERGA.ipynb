2. Aplicar la Normalziación a los conjuntos de datos


````python
# aplanar las 60.000 imágenes: de (28,28) a vectores de 784 features
# descartar features con varianza nula: (en MNIST muchas columnas de pixeles siempre son 0 : varianza = 0).
# normalizar (estandarizar): restar la media. dividir por la desviación estándar.

#defino la clase normalizar
class Normalizar:
    def __init__(self):
        self.mean = None
        self.std = None
        self.valid_idx = None  # índices de features que no tienen varianza nula

    # etapa de entrenamiento: calcula medias y desvios
    def fit(self, X):
        # calcula media y desviación estándar por feature
        mean = np.mean(X, axis=0) #media
        std = np.std(X, axis=0)   #varianza

        # guarda índices de columnas válidas (std > 0), descarta de features con varianza nula
        # varianza sero sería dividir por cero mas adelante, además no aporta info
        self.valid_idx = np.where(std > 0)[0]

        # guarda medias y varianzas solo de esas columnas
        self.mean = mean[self.valid_idx]
        self.std = std[self.valid_idx]

    # transforma datos usando lo calculado en fit, en este caso normaliza
    #esto va a servir mas adelante, teniendo los datos centraods y escalados facilita la convergencia del gradiente descendiente
    def transform(self, X):
        # mantiene solo columnas válidas
        X_valid = X[:, self.valid_idx]
        # Centrar y escalar
        return (X_valid - self.mean) / self.std

    # entrena y aplica en un solo paso
    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)

# 1. aplanar imágenes
X_train = x_train.reshape(x_train.shape[0], -1)  # (60000, 784)
X_test = x_test.reshape(x_test.shape[0], -1)    # (10000, 784)

print("Dimensiones antes de normalizar:", X_train.shape, X_test.shape)

# 2. normalizar
normalizador = Normalizar()
X_train_norm = normalizador.fit_transform(X_train)
X_test_norm = normalizador.transform(X_test)

print("Dimensiones después de normalizar:", X_train_norm.shape, X_test_norm.shape)
#esta bien que sea mas chico porque descarta valores

# verifico la media y varianza de las primeras 5 muestras
print("Media (train):", end="")
print(np.trunc(np.mean(X_train_norm, axis=0)[:5]).astype(int)) #grafica e^-20 asi que redondeo a cero
print("Varianza (train):", np.var(X_train_norm, axis=0)[:5])
````

(c) Análisis Teórico:
 1.  Calcular la función inversa $σ^{−1}(p)$ para $p ∈ (0,1)$, donde $σ(z)$ es la función sigmoide.
 2.  Sea $p = σ(z)$, calcular la derivada $σ′(z)$. Si el resultado se expresa en función de $p$, se simplifica
 bastante.
 3.  Hallar una expresión analítica para la función costo y su gradiente. Tenga en cuenta las expresiones asociadas a una regresión logística de dos clases.

### Función inversa de la sigmoide  

$$
p = \sigma(z) = \frac{1}{1+e^{-z}}
$$

$$
e^{-z} = \frac{1}{p} - 1 \quad \Rightarrow \quad z = \ln\left(\frac{p}{1-p}\right)
$$

$$
\sigma^{-1}(p) = \ln\left(\frac{p}{1-p}\right), \quad p \in (0,1)
$$


---

### Derivada de la sigmoide  

$$
\sigma'(z) = \frac{d}{dz}\left(\frac{1}{1+e^{-z}}\right) = \sigma(z)\,(1-\sigma(z))
$$

Expresada en función de $p = \sigma(z)$:  

$$
\sigma'(z) = p(1-p)
$$

---



### Función de costo y gradiente  

La función de costo (riesgo empírico) para la regresión logística binaria está dada por la *cross-entropy*:

$$
J(\mathbf{w}, b) = \frac{1}{n} \sum_{i=1}^n l(\mathbf{x}_i,Y_i) = -\frac{1}{n}\sum_{i=1}^n \big[ Y_i \log(\sigma(z_i)) + (1-Y_i)\log(1-\sigma(z_i)) \big], \quad z_i = \mathbf{w}^T\mathbf{x}_i + b
$$

Derivando la expresión del costo, y recordando que $z_i = \mathbf{w}^T\mathbf{x}_i + b$:

$$
\nabla J(\mathbf{w}, b) = -\frac{1}{n} \sum_{i=1}^n Y_i ~ \frac{1}{\sigma(z_i)} \cdot \sigma(z_i) \cdot (1 - \sigma(z_i)) \cdot \mathbf{x}_i + (1-Y_i) ~ \frac{1}{1 - \sigma(z_i)} (1 - \sigma(z_i)) (1-(1 - \sigma(z_i))) \cdot (-\mathbf{x}_i)
$$

$$
= -\frac{1}{n} \sum_{i=1}^n Y_i ~ (1 - \sigma(z_i)) ~ \mathbf{x}_i - (1-Y_i) ~ \sigma(z_i) ~ \mathbf{x}_i
$$

$$
=  -\frac{1}{n} \sum_{i=1}^n \Big[ Y_i - Y_i ~ \sigma(z_i) - \sigma(z_i) + Y_i ~ \sigma(z_i) \Big] \cdot \mathbf{x}_i
$$

$$
= -\frac{1}{n} \sum_{i=1}^n \Big[ Y_i - \sigma(z_i) \Big] \cdot \mathbf{x}_i
$$

Expresado como producto matricial, queda la expresión del **gradiente de la función de costo**:

$$
\nabla J(\mathbf{w}, b) = -\frac{1}{n}\, X^T \big( Y - \sigma(X\mathbf{w}+b) \big)
$$


(d) Regresión Logística:
1. Implementar una regresión logística binaria utilizando gradiente descendente. El código debe estar estructurado de la siguiente manera:


```
 class RegresionLogistica:

  # Etapa de entrenamiento
  def fit(self,X):

  # Etapa de testeo soft
  def predict_proba(self,X):
  
  # Etapa de testeo hard
  def predict(self,X):
  
  # Computar el Accuracy
  def accuracy(self,X,y):

  def cross_entropy(self,X,y):
```

2. Entrenar el algoritmo eligiendo un learning rate y una cantidad de iteraciones tal que el riesgo empírico de entrenamiento parezca converger. Graficar dicho riesgo en función del número de iteraciones.
3. Reportar el accuracy y la cross entropy de entrenamiento y testeo.

````python
class RegresionLogistica:
    def __init__(self, lr=0.1, n_iter=1000):
        self.lr = lr               # learning rate
        self.n_iter = n_iter       # cantidad de iteraciones
        self.w = None              # pesos
        self.b = None              # sesgo
        self.loss_history = []     # para guardar cross-entropy en cada iteración

    # función sigmoide
    def _sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    # etapa de entrenamiento
    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.n_classes = len(np.unique(y))
        self.loss_history = []

        if self.n_classes == 2:
            # caso binario
            self.w = np.zeros(n_features)
            self.b = 0
            for _ in range(self.n_iter):
                z = np.dot(X, self.w) + self.b
                y_pred = self._sigmoid(z)

                eps = 1e-15
                y_pred = np.clip(y_pred, eps, 1 - eps)  # estabilidad numérica
                loss = -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
                self.loss_history.append(loss)

                dw = (1 / n_samples) * np.dot(X.T, (y_pred - y))
                db = (1 / n_samples) * np.sum(y_pred - y)
                self.w -= self.lr * dw
                self.b -= self.lr * db
        else:
            # caso multiclase
            self.w = np.zeros((n_features, self.n_classes))
            self.b = np.zeros(self.n_classes)
            y_one_hot = np.eye(self.n_classes)[y]

            for _ in range(self.n_iter):
                z = np.dot(X, self.w) + self.b
                y_pred = self._softmax(z)

                eps = 1e-15
                y_pred = np.clip(y_pred, eps, 1 - eps)  # estabilidad numérica
                loss = -np.mean(np.sum(y_one_hot * np.log(y_pred), axis=1))
                self.loss_history.append(loss)

                dw = (1 / n_samples) * np.dot(X.T, (y_pred - y_one_hot))
                db = (1 / n_samples) * np.sum(y_pred - y_one_hot, axis=0)
                self.w -= self.lr * dw
                self.b -= self.lr * db

    def _softmax(self, z):
        exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))

        #normalizo para que la suma de probabilidades por fila sea 1
        #cada fila representa la distrib de prob sobre las clases para una muestra
        return exp_z / np.sum(exp_z, axis=1, keepdims=True)

    #probabilidades entre [0,1]
    def predict_proba(self, X):
        # Binario: devuelve shape (n_samples, 2), asi es consistente con el multiclase
        # Multiclase: devuelve shape (n_samples, n_classes)

        z = np.dot(X, self.w) + self.b  # (n_samples, n_classes) o (n_samples,)

        # Caso binario: sigmoid y convertimos a dos columnas [p0, p1]
        if z.ndim == 1 or z.shape[1] == 1:
            probs_pos = self._sigmoid(z)
            return np.vstack([1 - probs_pos, probs_pos]).T

        # Caso multiclase: softmax directo, tomo la clase con mayor probabilidad
        return self._softmax(z)

    #prediccion hard, binario revuelve [0,1], multiclase devuelve la clase con probabilidad máxima
    def predict(self, X):
        probs = self.predict_proba(X)
        return np.argmax(probs, axis=1)

    # accuracy
    # accuracy evalúa el porcentaje de aciertos, mientras que cross-entropy mide qué tan bien calibradas están las probabilidades.
    def accuracy(self, X, y):
        y_pred = self.predict(X)
        return np.mean(y_pred == y)

    # cross-Entropy (riesgo empírico), ajustado para multiclase
    def cross_entropy(self, X, y):
        y_pred = self.predict_proba(X)
        eps = 1e-15
        y_pred = np.clip(y_pred, eps, 1 - eps)

    # Binario (y en {0,1})
        if y_pred.shape[1] == 2:
            return -np.mean(y * np.log(y_pred[:, 1]) + (1 - y) * np.log(y_pred[:, 0]))

    # Multiclase (One-hot encoding implícito)
        n_samples = y.shape[0]
        return -np.mean(np.log(y_pred[np.arange(n_samples), y]))




# uso los datos ya normalizados
X_train_log = X_train_norm
X_test_log = X_test_norm
y_train_log = y_train_bin
y_test_log = y_test_bin

# entreno el modelo
model = RegresionLogistica(lr=0.1, n_iter=300)
model.fit(X_train_log, y_train_log)

# grafico el riesgo (cross-entropy) durante el entrenamiento
plt.plot(model.loss_history, color = "tomato")
plt.xlabel("Iteración")
plt.ylabel("Cross-Entropy")
plt.title("Evolución del riesgo en entrenamiento")
plt.grid()
plt.show()

# métricas
print("Entrenamiento:")
model_accuracy_train = model.accuracy(X_train_log, y_train_log)
print(" Accuracy:", model_accuracy_train)
model_cross_entropy_train = model.cross_entropy(X_train_log, y_train_log)
print(" Cross-Entropy:", model_cross_entropy_train)

print("\nTesteo:")
model_accuracy_test = model.accuracy(X_test_log, y_test_log)
print(" Accuracy:", model_accuracy_test)
model_cross_entropy_test = model.cross_entropy(X_test_log, y_test_log)
print(" Cross-Entropy:", model_cross_entropy_test)

#tiende a estabilizarse alrededor de 0.25
#el learning rate parece bien, si fuera muy alto la curva oscilaría; si fuera muy bajo, bajaría casi en línea recta y muy despacio
#no son necesarias 500 iteraciones
````

(e)  Curva ROC: A partir de la salida de predict_proba del conjunto de testeo, implementar la curva ROC.

````python
# La curva ROC compara: TPR (True Positive Rate) = sensibilidad = TP/TP+FN
# y FPR (False Positive Rate) = FP/FP+TN Para distintos umbrales en predict_proba.
#Si el umbral = 0 todo se clasifica como 1. Si el umbral = 1 todo se clasifica como 0.
#Al recorrer umbrales intermedios, obtenemos la curva ROC.

def curva_roc(y_true, y_proba, class_id=None):
    #Calcula los puntos para la curva ROC.
    #Para multiclase, calcula la curva "one-vs-rest" para la clase especificada por class_id.

    if y_proba.ndim == 2 and y_proba.shape[1] > 2:
        # multiclase: one-vs-rest
        if class_id is None:
            raise ValueError("Para multiclase, debe especificar 'class_id'.")
        y_true_binary = (y_true == class_id).astype(int)
        y_proba_binary = y_proba[:, class_id]
    else:
        # binario:
        y_true_binary = y_true # En binario y_true ya está en [0, 1]
        y_proba_binary = y_proba[:, 1]

    # ordeno por probabilidad descendente
    thresholds = np.linspace(0, 1, 100)  # 100 umbrales entre 0 y 1


    tpr_list = []
    fpr_list = []

    for thr in thresholds:
        # Clasificación hard según el umbral
        y_pred = (y_proba_binary >= thr).astype(int)

        # TP, FP, TN, FN
        TP = np.sum((y_true_binary == 1) & (y_pred == 1))
        FP = np.sum((y_true_binary == 0) & (y_pred == 1))
        TN = np.sum((y_true_binary == 0) & (y_pred == 0))
        FN = np.sum((y_true_binary == 1) & (y_pred == 0))

        # Calculo TPR y FPR. TPR: proporción de positivos bien
        # clasificados (sensibilidad). FPR: proporción de negativos mal clasificados.
        TPR = TP / (TP + FN) if (TP + FN) > 0 else 0
        FPR = FP / (FP + TN) if (FP + TN) > 0 else 0

        tpr_list.append(TPR)
        fpr_list.append(FPR)

    return np.array(fpr_list), np.array(tpr_list)


# obtengo probabilidades en test
y_proba_test = model.predict_proba(X_test_log)

# calculo puntos ROC - ahora para multiclase, especificando la clase (ej. clase 0)
# Puedes graficar una curva por cada clase, o promediar. Aquí graficamos para la clase 0 como ejemplo.
fpr, tpr = curva_roc(y_test_log, y_proba_test)

# grafico
plt.figure(figsize=(5,5))
plt.plot(fpr, tpr, label="ROC", color = "tomato")
plt.plot([0,1], [0,1], '--', label="Azar", color = "purple")  # diagonal, en todo momento hay 50/50 de tener falso positivo o positivo verdadero
plt.xlabel("False Positive Rate (FPR)")
plt.ylabel("True Positive Rate (TPR)")
plt.title("Curva ROC ")
plt.legend()
plt.grid()
plt.show()


#La curva está claramente arriba de la línea punteada (clasificador aleatorio). Esto significa que el modelo aprende a distinguir pares vs impares mejor que el azar.
#Área muy grande: La curva se acerca mucho al vértice superior izquierdo (FPR≈0, TPR≈1).Eso indica que para muchos umbrales, el modelo logra alta sensibilidad
# (TPR alto) con baja tasa de falsos positivos (FPR bajo)
#el clasificador funciona muy bien en este problema
````
